```python
import pandas as pd

# Load the dataset
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Select the relevant columns for analysis
columns_of_interest = [
    'Churn', 'Contract', 'tenure',
    'InternetService', 'PhoneService',
    'TechSupport', 'OnlineSecurity',
    'DeviceProtection',
    'PaperlessBilling', 'PaymentMethod',
    'TotalCharges'
]

df_filtered = df[columns_of_interest].copy()  # Create a copy to avoid SettingWithCopyWarning

# Convert 'Churn' to binary numeric for analysis
df_filtered.loc[:, 'Churn'] = df_filtered['Churn'].map({'No': 0, 'Yes': 1})

# Make sure 'TotalCharges' is numeric
df_filtered.loc[:, 'TotalCharges'] = pd.to_numeric(df_filtered['TotalCharges'], errors='coerce')
df.head()
```

```python
# Analyze the 'Churn' based on 'Contract' type
contract_analysis = df_filtered.groupby('Contract')['Churn'].mean().reset_index()
contract_analysis.columns = ['Contract', 'Churn Rate']
print("\nChurn Rate by Contract Type:")
print(contract_analysis)
```

```python
# Analyze 'Churn' based on 'tenure'
tenure_churn = df_filtered[['tenure', 'Churn']].groupby('tenure').mean().reset_index()
print("\nChurn Rate by Tenure in months:")
print(tenure_churn)
```

```python
# Analyze 'Churn' based on 'InternetService'
internet_service_analysis = df_filtered.groupby('InternetService')['Churn'].mean().reset_index()
internet_service_analysis.columns = ['Internet Service', 'Churn Rate']
print("\nChurn Rate by Internet Service:")
print(internet_service_analysis)
```

```python
# Analyze 'Churn' based on 'PhoneService'
phone_service_analysis = df_filtered.groupby('PhoneService')['Churn'].mean().reset_index()
phone_service_analysis.columns = ['Phone Service', 'Churn Rate']
print("\nChurn Rate by Phone Service:")
print(phone_service_analysis)
```

```python
# Analyze 'Churn' based on 'TechSupport'
tech_support_analysis = df_filtered.groupby('TechSupport')['Churn'].mean().reset_index()
tech_support_analysis.columns = ['Tech Support', 'Churn Rate']
print("\nChurn Rate by Tech Support:")
print(tech_support_analysis)
```

```python
# Analyze 'Churn' based on 'OnlineSecurity'
online_security_analysis = df_filtered.groupby('OnlineSecurity')['Churn'].mean().reset_index()
online_security_analysis.columns = ['Online Security', 'Churn Rate']
print("\nChurn Rate by Online Security:")
print(online_security_analysis)
```

```python
# Analyze 'Churn' based on 'DeviceProtection'
device_protection_analysis = df_filtered.groupby('DeviceProtection')['Churn'].mean().reset_index()
device_protection_analysis.columns = ['Device Protection', 'Churn Rate']
print("\nChurn Rate by Device Protection:")
print(device_protection_analysis)
```

```python
# Analyze 'Churn' based on 'PaperlessBilling'
paperless_billing_analysis = df_filtered.groupby('PaperlessBilling')['Churn'].mean().reset_index()
paperless_billing_analysis.columns = ['Paperless Billing', 'Churn Rate']
print("\nChurn Rate by Paperless Billing:")
print(paperless_billing_analysis)
```

```python
# Analyze 'Churn' based on 'TotalCharges' (as a numerical variable)
total_charges_analysis = df_filtered[['TotalCharges', 'Churn']].dropna()
correlation_total_charges = total_charges_analysis.corr().iloc[0, 1]  # Correlation between Churn and TotalCharges
print(f"\nCorrelation between Total Charges and Churn: {correlation_total_charges:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style
sns.set(style="whitegrid")

# Analysis of Churn based on Tenure
plt.figure(figsize=(10, 6))
sns.countplot(x='tenure', hue='Churn', data=df_filtered, palette='Set1')
plt.title('Churn Distribution by Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Count of Customers')
plt.legend(title='Churn', labels=['No Churn', 'Churn'])
plt.xticks(rotation=45)
plt.show()

# Analyze the average Total Charges for Churned vs. Not Churned
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='TotalCharges', data=df_filtered, palette='Set2')
plt.title('Total Charges by Churn Status')
plt.xlabel('Churn')
plt.ylabel('Total Charges')
plt.xticks(ticks=[0, 1], labels=['No Churn', 'Churn'])
plt.show()

# Correlation heatmap including Churn
numerical_features = df_filtered.select_dtypes(include=['number'])
correlation = numerical_features.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd

# Step 1: Define the states
states = ['Active', 'Churned']

# Step 2: Create the transition matrix
# Example transition probabilities
# Rows represent the current state, columns represent the next state
# P(Active to Active), P(Active to Churned), P(Churned to Active), P(Churned to Churned)
transition_matrix = np.array([[0.8, 0.2],  # Active
                              [0.3, 0.7]]) # Churned

# Step 3: Create a function to simulate the Markov Chain
def simulate_markov_chain(start_state, transition_matrix, n_steps):
    current_state = start_state
    states_history = [current_state]

    for _ in range(n_steps):
        if current_state == 'Active':
            current_state = np.random.choice(states, p=transition_matrix[0])
        else:
            current_state = np.random.choice(states, p=transition_matrix[1])
        states_history.append(current_state)

    return states_history

# Step 4: Simulate the Markov Chain for a specified number of steps
n_steps = 10  # For example, simulate for 10 periods
initial_state = 'Active'  # Start with an Active customer
result = simulate_markov_chain(initial_state, transition_matrix, n_steps)

# Step 5: Analyze the results
print("State transition history:")
print(result)

# Count occurrences of each state
state_counts = pd.Series(result).value_counts()
print("\nState counts:")
print(state_counts)

# Calculate the probabilities of being in each state after the simulation
state_probabilities = state_counts / len(result)
print("\nState probabilities:")
print(state_probabilities)
```
