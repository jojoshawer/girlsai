```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (ensure it's already uploaded in Colab)
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Preprocess the data
df['Churn'] = df['Churn'].map({'No': 0, 'Yes': 1})  # Map Churn to binary
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')  # Convert TotalCharges to numeric

# Chosen attributes for predictive model analysis
chosen_attributes = ['Contract', 'tenure', 'MonthlyCharges', 'TotalCharges']

# Analysis of chosen attributes
print("\n--- Analysis of Chosen Attributes for Predictive Model ---")

# 1. Contract Type Analysis
contract_analysis = df.groupby('Contract')['Churn'].mean().reset_index()
contract_analysis.columns = ['Contract', 'Churn Rate']
print("\nChurn Rate by Contract Type:")
print(contract_analysis)

# 2. Tenure Analysis
tenure_churn = df[['tenure', 'Churn']].groupby('tenure').mean().reset_index()
print("\nChurn Rate by Tenure in Months:")
print(tenure_churn)

# 3. Monthly Charges Analysis
monthly_charges_analysis = df[['MonthlyCharges', 'Churn']].dropna()
correlation_monthly_charges = monthly_charges_analysis.corr().iloc[0, 1]
print(f"\nCorrelation between Monthly Charges and Churn: {correlation_monthly_charges:.4f}")

# 4. Total Charges Analysis
total_charges_analysis = df[['TotalCharges', 'Churn']].dropna()
correlation_total_charges = total_charges_analysis.corr().iloc[0, 1]
print(f"\nCorrelation between Total Charges and Churn: {correlation_total_charges:.4f}")

# Visualization of Chosen Attributes
plt.figure(figsize=(14, 8))

# Contract Type Plot
plt.subplot(2, 2, 1)
sns.barplot(data=contract_analysis, x='Contract', y='Churn Rate', palette='viridis')
plt.title('Churn Rate by Contract Type')

# Tenure Distribution Plot
plt.subplot(2, 2, 2)
sns.histplot(df, x='tenure', hue='Churn', multiple='stack', palette='muted')
plt.title('Distribution of Tenure by Churn')

# Monthly Charges KDE Plot
plt.subplot(2, 2, 3)
sns.kdeplot(df['MonthlyCharges'][df['Churn'] == 0], label="Not Churn", shade=True, color="blue")
sns.kdeplot(df['MonthlyCharges'][df['Churn'] == 1], label="Churn", shade=True, color="red")
plt.title('Distribution of Monthly Charges by Churn')
plt.legend()

# Total Charges KDE Plot
plt.subplot(2, 2, 4)
sns.kdeplot(df['TotalCharges'][df['Churn'] == 0], label="Not Churn", shade=True, color="blue")
sns.kdeplot(df['TotalCharges'][df['Churn'] == 1], label="Churn", shade=True, color="red")
plt.title('Distribution of Total Charges by Churn')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import entropy

# Load dataset
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Preprocess the data
df['Churn'] = df['Churn'].map({'No': 0, 'Yes': 1})
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')  # Convert TotalCharges to numeric

# List of chosen attributes for the analysis
chosen_attributes = ['Contract', 'tenure', 'MonthlyCharges', 'TotalCharges']

### 1. Convert Categorical Variables to Numerical (One-Hot Encoding)
df_encoded = pd.get_dummies(df[chosen_attributes], drop_first=True)
df_encoded['Churn'] = df['Churn']  # Include the target variable in the encoded DataFrame

### 2. Calculate Correlation with Churn (for numerical and one-hot encoded attributes)
print("\n--- Correlation with Churn ---")
correlations = df_encoded.corr()['Churn'].drop('Churn')
print(correlations)

### 3. Calculate Entropy (for categorical attributes)

# Function to calculate entropy of a categorical feature with respect to churn
def calculate_entropy(feature, target):
    contingency_table = pd.crosstab(df[feature], df[target])
    entropy_value = entropy(contingency_table.sum(axis=1), base=2)
    return entropy_value

print("\n--- Entropy with Churn ---")
# Apply entropy calculation for categorical attributes
categorical_attributes = ['Contract']  # Contract is the only categorical attribute in chosen ones

for attribute in categorical_attributes:
    entropy_value = calculate_entropy(attribute, 'Churn')
    print(f"Entropy of {attribute} with respect to Churn: {entropy_value:.4f}")
```

```python

import numpy as np
from hmmlearn import hmm


states = ["Low Risk", "Moderate Risk", "High Risk"]
n_states = len(states)
print('Number of hidden states:', n_states)

observations = ["Churn_No", "Churn_Yes"]
n_observations = len(observations)
print('Number of observations:', n_observations)

state_probability = np.array([0.5, 0.3, 0.2])
print("State probability:", state_probability)

transition_probability = np.array([
    [0.7, 0.2, 0.1],
    [0.3, 0.5, 0.2],
    [0.2, 0.3, 0.5]
])
print("\nTransition probability:\n", transition_probability)

emission_probability = np.array([
    [0.9, 0.1],
    [0.6, 0.4],
    [0.3, 0.7]
])
print("\nEmission probability:\n", emission_probability)

model = hmm.CategoricalHMM(n_components=n_states)
model.startprob_ = state_probability
model.transmat_ = transition_probability
model.emissionprob_ = emission_probability

observations_list = [0, 0, 1, 1, 0, 1, 0]
observations_sequence = np.array(observations_list).reshape(-1, 1)
print('Observations sequence:', ' '.join([observations[x] for x in observations_list]))

hidden_states = model.predict(observations_sequence)
hidden_states_text = ' '.join([states[x] for x in hidden_states])
print("Most likely hidden states:", hidden_states_text)

log_probability, hidden_states = model.decode(observations_sequence, algorithm='viterbi')
print('Log Probability:', log_probability)
print('Most likely hidden states (Viterbi):', ' '.join([states[x] for x in hidden_states]))
```

```python
# Import necessary libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set up figure for subplots
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Transition Probability Heatmap
sns.heatmap(transition_probability, annot=True, cmap="Blues", fmt=".2f",
            xticklabels=states, yticklabels=states, ax=ax[0])
ax[0].set_title("Transition Probabilities")
ax[0].set_xlabel("To State")
ax[0].set_ylabel("From State")

# Emission Probability Heatmap
sns.heatmap(emission_probability, annot=True, cmap="Greens", fmt=".2f",
            xticklabels=observations, yticklabels=states, ax=ax[1])
ax[1].set_title("Emission Probabilities")
ax[1].set_xlabel("Observations (Churn Status)")
ax[1].set_ylabel("Risk State")

# Display the heatmaps
plt.tight_layout()
plt.show()
```
