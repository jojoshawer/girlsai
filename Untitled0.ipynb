```python
import pandas as pd
import numpy as np
from scipy.stats import entropy
from sklearn.feature_selection import mutual_info_classif

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

df['Churn'] = df['Churn'].map({'No': 0, 'Yes': 1})
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())


chosen_attributes = ['Contract', 'tenure', 'MonthlyCharges', 'TotalCharges']


df_encoded = pd.get_dummies(df[chosen_attributes], drop_first=True)
df_encoded['Churn'] = df['Churn']

print("\n--- Correlation with Churn ---")
correlations = df_encoded.corr()['Churn'].drop('Churn')
print(correlations)


def calculate_entropy(feature, target):
    contingency_table = pd.crosstab(df[feature], df[target])
    entropy_value = entropy(contingency_table.sum(axis=1), base=2)
    return entropy_value

print("\n--- Entropy with Churn ---")
categorical_attributes = ['Contract']

for attribute in categorical_attributes:
    entropy_value = calculate_entropy(attribute, 'Churn')
    print(f"Entropy of {attribute} with respect to Churn: {entropy_value:.4f}")

features = df_encoded.drop('Churn', axis=1)
target = df_encoded['Churn']
mutual_info = mutual_info_classif(features, target)
print(mutual_info)
```

```python
import numpy as np
import pandas as pd


df_filtered['Churn_Lag'] = df_filtered['Churn'].shift(1)

df_filtered = df_filtered.dropna(subset=['Churn_Lag'])

transition_matrix = pd.crosstab(df_filtered['Churn_Lag'], df_filtered['Churn'], normalize='index')

transition_matrix.columns = ['Churned', 'Not Churned']
transition_matrix.index = ['Churned (previous)', 'Not Churned (previous)']

print("Transition Matrix:")
print(transition_matrix)


current_state = [0, 1]

next_state_probabilities = np.dot(current_state, transition_matrix)

print("\nNext State Probabilities (from 'Not Churned'):")
print(f"Probability of Churning Tomorrow: {next_state_probabilities[1]:.4f}")
print(f"Probability of Not Churning Tomorrow: {next_state_probabilities[0]:.4f}")
```

```python
import numpy as np
steps = 5

state_vector = initial_state_prob

# Calculate the distribution over time
for i in range(steps):
    state_vector = np.dot(state_vector, transition_matrix)
    print(f"State distribution after {i+1} months: Not Churned: {state_vector[0]:.4f}, Churned: {state_vector[1]:.4f}")
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

transition_matrix = np.array([
    [0.7394, 0.2606],
    [0.2672, 0.7328]
])
initial_state_prob = np.array([0.7346, 0.2654])


G = nx.DiGraph()
states = ["Not Churned", "Churned"]

for i, state_from in enumerate(states):
    for j, state_to in enumerate(states):
        G.add_edge(state_from, state_to, weight=transition_matrix[i, j])


plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=12, font_weight="bold")
edge_labels = {(states[i], states[j]): f"{transition_matrix[i, j]:.2f}" for i in range(2) for j in range(2)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red")
plt.title("Markov Model Transition Diagram")
plt.show()


steps = 5
state_probabilities = [initial_state_prob]
current_state = initial_state_prob

for _ in range(steps):
    next_state = np.dot(current_state, transition_matrix)
    state_probabilities.append(next_state)
    current_state = next_state


state_df = pd.DataFrame(state_probabilities, columns=states)
state_df.index.name = "Steps"

plt.figure(figsize=(10, 6))
plt.plot(state_df.index, state_df["Not Churned"], label="Not Churned", marker='o', color='blue')
plt.plot(state_df.index, state_df["Churned"], label="Churned", marker='o', color='orange')
plt.xlabel("Time Steps")
plt.ylabel("Probability")
plt.title("Probability Evolution of Customer Churn States Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
